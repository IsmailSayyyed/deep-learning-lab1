```python
import numpy as np
import pandas as pd
import os

from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

iris=load_iris()
samples=iris.data
model=KMeans(n_clusters=3)
model.fit(samples)
labels=model.predict(samples)
print(labels)
xs=samples[:,0]
ys=samples[:,2]
plt.scatter(xs,ys,c=labels)
centroids=model.cluster_centers_
centroids_x=centroids[:,0]
centroids_y=centroids[:,2]
plt.scatter(centroids_x,centroids_y,marker='D',s=50)
plt.show()
from google.colab import files
uploaded=files.upload()
import io
df = pd.read_csv(io.BytesIO(uploaded['Housing.csv']))
print(df)
dataset = df.values
dataset
b=range(0,4)
PTRATIO =[]
for k in b:model=KMeans(n_clusters=k)
plt.plot(b,'-h')
plt.xlabel('Number of Clusters , k')
plt.ylabel('PTRATIO ')
plt.xticks(b)
plt.show()
df.head()
model=KMeans(n_clusters=3)
centroids
xs_ZN=df.iloc[:,0]
ys_AGE=df.iloc[:,-1]
centroids_xs_ZN=centroids[:,0]
centroids_ys_AGE=centroids[:,-1]
plt.scatter(xs_ZN,ys_AGE)
plt.scatter(centroids_xs_ZN,centroids_ys_AGE,c='red')
plt.show()
```
